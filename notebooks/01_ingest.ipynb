```python
import pandas as pd
import sqlite3
from pathlib import Path

RAW = Path("../data/raw")
DB_PATH = Path("../data/sqlite/retail.db")
DB_PATH.parent.mkdir(parents=True, exist_ok=True)

conn = sqlite3.connect(DB_PATH)
```

```python
order_products = pd.read_csv(RAW / "retail_transactions.csv")

order_products.shape
```

```python
order_products.to_sql("order_products", conn, if_exists="replace", index=False)

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect(DB_PATH)
pd.read_sql("SELECT COUNT(*) AS n_rows FROM order_products;", conn)
```

```python
pd.read_sql("SELECT * FROM order_products LIMIT 5;", conn)
```
