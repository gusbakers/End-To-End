```python
import pandas as pd
import sqlite3
from pathlib import Path

DB_PATH = Path("../data/sqlite/retail.db")
PROCESSED = Path("../data/processed")
PROCESSED.mkdir(parents=True, exist_ok=True)

conn = sqlite3.connect(DB_PATH)

order_products = pd.read_sql("SELECT * FROM order_products;", conn)

order_products.head()
```

```python
order_products_clean = order_products.copy()

# reordered column cleaning
order_products_clean["reordered"] = (
    order_products_clean["reordered"]
    .fillna(0)
    .astype(int)
)

# tipos (defensivo)
order_products_clean["product_id"] = order_products_clean["product_id"].astype(int)
order_products_clean["quantity"] = order_products_clean["quantity"].astype(int)

order_products_clean.head()
```

```python
order_products_clean.to_csv(PROCESSED / "order_products_clean.csv", index=False)

print("Data saved successfully to:")
print(f"  - {PROCESSED / 'order_products_clean.csv'}")
```

```python
order_products_clean.to_sql("order_products_clean", conn, if_exists="replace", index=False)

print("Cleaned data saved to database:")
print(f"  - Table: order_products_clean")

conn.close()
```
